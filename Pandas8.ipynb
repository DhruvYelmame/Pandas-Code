```python
df1 = pd.DataFrame({'RollNo':[1,2,3,4,5], 'Physics':[51,66,46,87,28]})
df1
```

```python
df3 = pd.DataFrame({'RollNo':[1,2,3,4,5], 'Physics':[33,62,42,69,45]})
df4 = pd.DataFrame({'RollNo':[1,2,3,4,5], 'Chemistry':[33,62,42,69,45]})
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4, how='left')
```

```python
pd.merge(df3,df4, how='right')
```

```python
pd.merge(df3,df4, how='outer')
```
